```python
import json
import numpy as np
import matplotlib.pyplot as plt

#
depth_map = np.load('output/20241130_121251297_iOS.npy')
depth_map = np.squeeze(depth_map)  # Removes dimensions of size 1
depth_map = depth_map[:, :, 0]  # Select only the first channel

# Display the depth map
plt.imshow(depth_map, cmap='viridis')  # 'inferno' is just an example colormap, you can use others like 'gray', 'viridis', etc.
plt.colorbar(label='Depth Value')  # Adds a color bar to indicate depth values
plt.title('Depth Map Visualization')
plt.show()
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt

#
with open('output/20241130_121251297_iOS.json', 'r') as file:
    depth_data = json.load(file)
depth_map = np.array(depth_data)
depth_map = np.squeeze(depth_map)  # Removes dimensions of size 1
depth_map = depth_map[:, :, 0]  # Select only the first channel

# Display the depth map
plt.imshow(depth_map, cmap='viridis')  # 'inferno' is just an example colormap, you can use others like 'gray', 'viridis', etc.
plt.colorbar(label='Depth Value')  # Adds a color bar to indicate depth values
plt.title('Depth Map Visualization')
plt.show()
```
